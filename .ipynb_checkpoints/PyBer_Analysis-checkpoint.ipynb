```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
# Get the sum of the fares for each city type.
total_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_type
```

```python
# Get the sum fo the drivers by city type.
total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_type
```

```python
# Get the number of rides by city type.
total_rides_by_type= pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type
```

```python
# Find the average fare per ride for each city type.
avg_fare_ride= total_fares_by_type/total_rides_by_type
avg_fare_ride
```

```python
# Find the average fare per driver for each city type.
avg_fare_driver= total_fares_by_type/ total_drivers_by_type
avg_fare_driver
```

```python
# Create a summary data frame.
pyber_summary_df = pd.DataFrame(
    {"Total Rides": total_rides_by_type,
        "Total Drivers": total_drivers_by_type,
        "Total Fares": total_fares_by_type,
        "Average Fare per Ride": avg_fare_ride,
        "Average Fare per Driver": avg_fare_driver})
```

```python
#Format the summary data frame.

pyber_summary_df["Total Rides"]=pyber_summary_df["Total Rides"].map("{:,}".format)

pyber_summary_df["Total Drivers"]=pyber_summary_df["Total Drivers"].map("{:,}".format)

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
```

```python
# Remove the index title. 
pyber_summary_df.index.name= None
```

```python
pyber_summary_df
```

```python
# Rename the columns in the Pyber DF.
pyber_rename= pyber_data_df.rename(columns=
{'city':'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id','driver_count': 'No. Drivers', 'type':'City Type'})
pyber_rename
```

```python
# Set the Date as the index column.
pyber_rename.set_index(["Date"])
```

```python
# Name columns to copy 
columns= pyber_rename[["Date", "City Type","Fare"]]
```

```python
# Copy columns into new DF.
pyber_fares_df= columns.copy()
```

```python
# Convert the Date column to datetime format
pyber_fares_df["Date"]=pd.to_datetime(pyber_fares_df["Date"])
```

```python
pyber_fares_df.info()
```

```python
pyber_fares_df
```

```python
# Set the date as the index for new DF.
pyber_fares_df.set_index(["Date"])
```

```python
# Get the sum of the fares for each city type.
pyber_fares_df = pyber_fares_df.groupby(["City Type", "Date"]).sum()["Fare"]
pyber_fares_df
```

```python
# Convert the Series to a Data Frame
pyber_fares_df=pyber_fares_df.to_frame()
pyber_fares_df
```

```python
#Create a new DF with the correct headings
fares_table_df=pd.pivot_table(pyber_fares_df, values="Fare", index=["Date"], columns=["City Type"])
```

```python
fares_table_df
```

```python
#Select only the year 2019
dates_df= fares_table_df.loc['2019-01-01':'2019-04-28']
```

```python
dates_df
```

```python
#Sort the dates by weeks
weekly_resampled_data = dates_df.resample('W').sum()
```

```python
weekly_resampled_data
```

```python
%matplotlib inline
```

```python
#Create the graph
plt.style.use('fivethirtyeight')

weekly_resampled_data.plot(figsize = (14,4))
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($)")
plt.xlabel("")
plt.grid(True)
plt.legend(bbox_to_anchor=[0.5, 0.5], 
           loc='center', ncol=1)
# Save the figure.
plt.savefig("analysis/ChallengeFig.png")
plt.show()
```
